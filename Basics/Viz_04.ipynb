## **Pyplot vs Axes**

| **Funcionalidad** | **`plt` (Pyplot)**    | **`ax` (Axes)**      |
|-------------------|-----------------------|----------------------|
| Título            | `plt.title()`         | `ax.set_title()`     |
| Etiqueta X        | `plt.xlabel()`        | `ax.set_xlabel()`    |
| Etiqueta Y        | `plt.ylabel()`        | `ax.set_ylabel()`    |
| Límites X         | `plt.xlim()`          | `ax.set_xlim()`      |
| Límites Y         | `plt.ylim()`          | `ax.set_ylim()`      |
| Escala X          | `plt.xscale()`        | `ax.set_xscale()`    |
| Escala Y          | `plt.yscale()`        | `ax.set_yscale()`    |
| Ticks X           | `plt.xticks()`        | `ax.set_xticks()`    |
| Ticks Y           | `plt.yticks()`        | `ax.set_yticks()`    |
| Grilla            | `plt.grid()`          | `ax.grid()`          |
| Leyenda           | `plt.legend()`        | `ax.legend()`        |
| Ejes              | `plt.axis()`          | `ax.axis()`          |
| Ajustar layout    | `plt.tight_layout()`  | `fig.tight_layout()` |
| Añadir texto      | `plt.text()`          | `ax.text()`          |
| Anotación         | `plt.annotate()`      | `ax.annotate()`      |
| Colorbar          | `plt.colorbar()`      | `fig.colorbar()`     |
| Dibujar línea     | `plt.plot()`          | `ax.plot()`          |
| Dibujar scatter   | `plt.scatter()`       | `ax.scatter()`       |
| Dibujar barras    | `plt.bar()`           | `ax.bar()`           |
| Dibujar histograma| `plt.hist()`          | `ax.hist()`          |
| Dibujar imagen    | `plt.imshow()`        | `ax.imshow()`        |



```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv")
df.head()
```

```python
df.set_index("Country", inplace=True)
df.head()
```

```python
df.shape
```

# **Line Plot**
---

```python
years = list(map(str, range(1980, 2014)))
```

```python
df_line = df[years]

total_inmigrants = df_line.sum()
total_inmigrants
```

```python
fig, ax = plt.subplots()

ax.plot(total_inmigrants)

ax.set_title("Inmigrants between 1980 to 2013")

ax.set_xlabel("Years")
ax.set_ylabel("Total Inmigrants")

plt.show()
```

EL eje de las X se ve mal pues los años son de tipo 'str'.

```python
fig, ax = plt.subplots()

total_inmigrants.index = total_inmigrants.index.map(int)

ax.plot(total_inmigrants)

ax.set_title("Inmigrants between 1980 to 2013")

ax.set_xlabel("Years")
ax.set_ylabel("Total Inmigrants")

plt.show()
```

### Ahora vamos a modificarlo

```python
fig, ax = plt.subplots()

ax.plot(total_inmigrants,
       marker= 's',
       markersize=5,
       color='green',
       linestyle='dotted')

ax.set_title("Inmigrants between 1980 to 2013")

ax.set_xlabel("Years")
ax.set_ylabel("Total Inmigrants")

plt.xlim(1975, 2015)

plt.grid(True)

plt.show()
```

### Inmigración desde Haití

```python
df_haiti = df.loc[["Haiti"], years]
df_haiti = df_haiti.transpose()
df_haiti.head()
```

# **Scatter Plot**
---

```python
fig,ax = plt.subplots(figsize=(8, 4))

ax.scatter(list(map(int, years)), total_inmigrants,
          marker='o',
          s=20,
          color='darkblue')

plt.title("Inmigraciones entre 1980 a 2013")

plt.xlabel("Años")
plt.ylabel("Total de inmigraciones")

plt.grid(True)

ax.legend(["Inmigrants"], loc='upper center')

plt.show()
```

# **Histogram**

```python
df_country = df.groupby(["Country"])["2013"].sum().reset_index()
df_country.head()
```

```python
#Create figure and axes
fig, ax = plt.subplots(figsize=(10, 4))
ax.hist(df_country['2013'])
ax.set_title('New Immigrants in 2013') 
ax.set_xlabel('Number of Immigrants')
ax.set_ylabel('Number of Countries')
ax.legend(['Immigrants'])

#Display the plot
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
count = ax.hist(df_country['2013'])

#you can check the arrays in count with indexing count[0] for count, count[1] for bins

ax.set_title('New Immigrants in 2013') 
ax.set_xlabel('Number of Immigrants')
ax.set_ylabel('Number of Countries')
ax.set_xticks(list(map(int,count[1])))
ax.legend(['Immigrants'])

#Display the plot
plt.show()
```

# **Sub-plotting**
---

```python
fig,axs = plt.subplots(1, 2, sharey=True)

axs[0].plot(total_inmigrants)
axs[0].set_title("Line Plot on inmigrants")

axs[1].scatter(total_inmigrants.index, total_inmigrants)
axs[1].set_title("Scatter plot on inmigrants")

axs[0].set_ylabel("Cantidad de inmigrantes")

fig.suptitle("Ejemplo de Subplotting", fontsize=15)

fig.tight_layout()

plt.show()
```

Los siguinetes tres métodos son equivalentes:
* `ax = fig.add_subplot(1, 1, 1)`
* `ax = plt.subplot(1, 1, 1)`
* `fig, ax = plt.subplots()`

```python
# Create a figure with Four axes - two rows, two columns
fig = plt.figure(figsize=(8,4))

# Add the first subplot (top-left)
axs1 = fig.add_subplot(1, 2, 1)
#Plotting in first axes - the left one
axs1.plot(total_inmigrants)
axs1.set_title("Line plot on immigrants")

# Add the second subplot (top-right)
axs2 = fig.add_subplot(1, 2, 2)
#Plotting in second axes - the right one
axs2.barh(total_inmigrants.index, total_inmigrants) #Notice the use of 'barh' for creating horizontal bar plot
axs2.set_title("Bar plot on immigrants")
            
#Adding a Title for the Overall Figure
fig.suptitle('Subplotting Example', fontsize=15)

# Adjust spacing between subplots
fig.tight_layout()


# Show the figure
plt.show()
```
