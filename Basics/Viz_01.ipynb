# **Tipos de gráficas**
---
**Line plot o Line chart**
* Series de puntos conectados entre líneas.
* Ilustra relación entre dos eventos.
* Sirve para visualización a lo largo del tiempo.

**Bar plot o Bar chart**
* Comparar diferentes categorías o grupos

**Scatter plot**
* Examinar relación entre dos variables continuas
* Investigar patrones o tendencias en datos.

**Box plot**
* Comparar la distribución de variables continuas a través de diferentes categorías o grupos.
* Comparar distribuciones de múltiples variables.

**Histogram**
* Representa la distribución historíca de una variable.
* Comprender la distribución de los datos.

## **Matplotlib**
---
* `plot` para serie de puntos conectados entre lineas.
* `bar` for vertical bar plots
* `barh` for horizontal bar plots
* `hist` for histogram
* `box` for boxplot
* `kde` or `density` for density plots

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
#%matplotlib notebook
```

```python
print("Matplotlib version: ", mpl.__version__)
```

```python
haiti = df.loc["Haiti", years]
haiti.head()
```

```python
haiti.plot()
```

```python
haiti.plot()

plt.title("Inmigraciones desde Haití")
plt.ylabel("Número de inmigraciones")
plt.xlabel("Años")

plt.show()
```

Hay que tener en cuenta que los años (*Years*) son de tipo string, así que hay que cambiarlos a int primero

```python
haiti.index = haiti.index.map(int)
haiti.plot(kind='line') # Se puede definir el tipo de plot

plt.title("Inmigraciones desde Haití")
plt.ylabel("Número de inmigraciones")
plt.xlabel("Años")

plt.text(2000, 6100, "Terremoto de 2010")

plt.show()
```

NOTA: dado que "years" es de tipo entero, se especifica x como tal (*2000*), sin embargo, si "years" fuese de tipo string, se debería especificar el ínidice al que pertenece el valor al que queremos hacer referencia (en este caso 20).

```python
ci = df.loc[["China", "India"], years]
ci.head()
```

```python
ci.plot(kind='line')
```

```python
ci = ci.transpose()
```

```python
ci.plot(kind='line')

plt.title("Inmigraciones desde China e India")
plt.xlabel("Años")
plt.ylabel("Cantidad de inmigraciones")

plt.show()
```

### Comparar la tendencia de los 5 paises que contribuyeron más a la inmigración a Canadá

```python
inplace=True

df.sort_values(by='Total', ascending=False, axis=0, inplace=True)

df.head()
```

```python
df_top5 = df.head() # Quedarnos solo con los primeros 5 valores

df_top5 = df_top5[years].transpose()

df_top5.head()
```

```python
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='line', figsize=(14, 8))

plt.title("Tendencia de Inmigraciones de los 5 paises top")
plt.ylabel("Número de inmigraciones")
plt.xlabel("Años")

plt.show()
```
